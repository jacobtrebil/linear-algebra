```python
import numpy as np
from scipy import linalg
```

```python
A = np.array([[1, 1, 1], [1, 2, 4], [1, 2, 4]])
b = np.array([5, 13, 25]).reshape((3, 1))
```

```python
p, *_ = linalg.lstsq(A, b)
p
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 3, 1000)
y = p[0] + p[1] * x + p[2] * x ** 2
```

```python
plt.plot(x, y)
plt.plot(1, 5, "ro")
plt.plot(2, 13, "ro")
plt.plot(2, 25, "ro")
```
